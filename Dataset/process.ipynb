```python
import pandas as pd

train = pd.read_excel(r"D:\UIT\NCKH\Source Code\Demo\Dataset\Train.xlsx")
dev = pd.read_excel(r"D:\UIT\NCKH\Source Code\Demo\Dataset\Dev.xlsx")
test = pd.read_excel(r"D:\UIT\NCKH\Source Code\Demo\Dataset\Test.xlsx")

df = pd.concat([train, dev, test], ignore_index=True)
```

```python
# Ép toàn bộ column object về string
for col in df.columns:
    if df[col].dtype == "object":
        df[col] = df[col].astype(str)
```

```python
print('số lượng mẫu trong df: ', df.shape[0])
```

```python
df_latest = df.loc[df.groupby("Premise")["Law Date"].idxmax()]
print('số lượng mẫu trong df_latest: ', df_latest.shape[0])
```
